Q: What do 75%, 50%, 25% and 10% probabilities and associated $$ amounts mean?

 

These probabilities are based on the ACTUAL RESULTS and not approximations; they are the facts based on our 54 months backtesting. 

 

For example, let's look at this specific 5-min-bar-LONG PPA to illustrate the point:

 

long_10pct_conf: $260.18

long_25pct_conf: $139.05

long_50pct_conf: $58.71

long_75pct_conf: $16.00

 

This example means that out of all 5min-bar-LONG PPAs generated over the last 54 months,

25% exceeded 139.05_dollars and 10% exceeded $260.18 gain.  

Keep in mind that these probabilities are applicable ONLY while the PPA is active, which means that a vast number of the BTC trading chart conditions determined by our Cortical AI algorithm must be met before a PPA can be generated. During this time frame there were 11,393 5min-bar SHORT PPAs.

# PPAs in context: truffle hunting
- This isn't a "what will BTC do next at any given moment?" model
- This is a "Occasionally the HTM net picks up on certain patterns and sends out certain alerts" model. 
- The premise is "alert = certain pattern occuring -> here's what will probably happen"
    - Every alert theoretically makes the system 'smarter', as a bonus of HTM functionality
    - specifically, long (increase) or short (decrease) alerts for 5-min, 30-min and 60-min alerts, which respectively cover 1.25-hour, 7.5-hour and 15-hour periods.
        - 5-min alerts can trigger every 5 minutes, sort of thing.
        - a 25% confidence of 139 on a 30-minute-long:
            - during all 30-minute-longs in the past, 25 percent of them saw >$139 rises in BTC price.

## Pigs and Cortical Algorithms
If you and your mate take a pig sniffing for truffles, it's not going to find a truffle under every tree, because that's not how truffles work.

Every now and then, the beast's porcine snuffling catches a whiff of 'something' - perhaps the scent of a truffle, perhaps something else. Pigs aren't as precise as cortical algorithms. It starts digging and finds either a truffle, something else or nothing at all. You give it some scraps of your lunch, which it eats happily regardless of what you had for lunch, because it's a pig.

The HTM net / cortical model is a pig sniffing for 'alerts'. If it catches a whiff of '5-minute-long-alert', then it thinks BTC will drop while the short is active (minutes times 15 bars, so 75 minutes). So our 'active_frame' for this 5-minute-long-alert is 75 minutes.

The model then looks back at all the 5-minute-long-alerts it's called in the past, and checks how real BTC prices changed during the active_frames of those alerts. Let's say this particular alert has a 75% confidence of $16. 
During the active_frames of 75% of all past 5-minute-long-alerts, BTC went up at least 16 dollars.

## Gambling

To continue the metaphor, we must add gambling into the mix. As soon as the pig seems to catch the 5-minute-long scent, you bet your mate that there's a truffle worth $X beneath the tree. If there's a truffle they pay you X, if not you pay them X.

We can equate 'the pig digging through soil' with 'waiting while the real BTC price changes' - at the end we see our prize, whatever it may be. Let's say there was indeed a truffle worth \\$50.  Your mate pays you $50 (even though it's their pig. perhaps they keep the truffle?).

During the 75-minute active_frame for the 5-minute-long alert, BTC goes up \\$50. You sell one bitcoin and you're now \\$50 richer than before.

- If this doesn't happen, of course, you lose money on the bet and curse the malodorous swine that brought upon you such ruin. Ah, the joys of speculation.

# "why should you trust this HTM net?"
- If you buy the longs and sell the shorts, amount varying by the value of 75\\% confidence, then you'll make money in the long run. Why?
- To make money off of this, there's 2 assumptions that need to be true:
    - Internal Consistency: The HTM net is consistent with itself: 
        - a 5-minute-long will be triggered by a 'similar enough' set of stimulus/patterns/conditions as past 5-minute-longs
    - External Consistency: The predicted alerts are consistent with real-world price change:
        - 5-minute-long alerts will, _in general,_ predict a >\\$0 increase in price _over a 75-minute period_.
## Double-consistent trading
- Internal consistency here is equivalent to "If you give this cat-recognition convolutional net a cat, what are the odds it'll say 'cat'?"
    - I've done enough work with HTM nets to say that they're bloody good with patterns and quite picky depending on how you configure them. I'll put my vote in for the first assumption.
- External consistency is trickier: How can we be sure actual BTC price will change according to the predicted alert?
    - We can't - _this time_. But we _can_ look back at previous alerts and see how well they did. If 75% of 30-minute-shorts predicted \\$50 drops, then we can say "there's a 75% BTC will drop by at least \\$50 in the next 7.5 hours" - _if_ we believe in internal consistency.

## Example of _'loadsamoney'_ strategy
- Every time you get a 5-minute-long, buy \\$X worth of bitcoin. Wait 75 minutes. Sell \\$X worth, even if the price went down a bit (Or hold that coin until it goes above Price(time=0).
- There's a 75% chance that the price went up - on average for 5-min-longs, at least \\$19.
    - the other 25% includes values less than the predicted (usually around 19), or even negatives / price drops on the predicted long.
- This is significantly above 50, so if internal and external consistency holds, then you'll make _loadsamoney_ in the long run.

- To improve the strategy, perhaps hold out for longer alerts (larger time rangers / active_frames have more room for greater price changes) to cover trading costs, and trade alongside the other 5 of 6 'scents' the pig sniffs out.
